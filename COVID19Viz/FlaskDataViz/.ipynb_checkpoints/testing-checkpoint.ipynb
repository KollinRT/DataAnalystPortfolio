```python
import pandas as pd

# assuming your data is in a DataFrame called 'df'

# convert 'Week Ending Date' column to datetime type
df['Week Ending Date'] = pd.to_datetime(df['Week Ending Date'])

# set the 'Week Ending Date' column as the index
df.set_index('Week Ending Date', inplace=True)

# group the data by state and resample by month
df_monthly = df.groupby('Jurisdiction of Occurrence').resample('M').sum()

# reset the index to make 'Jurisdiction of Occurrence' and 'Week Ending Date' columns
df_monthly = df_monthly.reset_index()
```

# Now lets showcase the monthly COVID data each in its own CSV

```python
state_month_combinations = df_monthly[['Jurisdiction of Occurrence', 'Week Ending Date']].drop_duplicates()
```

```python
state_month_combinations
```

```python
# assuming you already have the 'df_monthly' DataFrame from the previous example

# get unique combinations of state and month
state_month_combinations = df_monthly[['Jurisdiction of Occurrence', 'Week Ending Date']].drop_duplicates()

# iterate over the combinations and save each subset of data into a separate .csv file
for state, month in state_month_combinations.itertuples(index=False):
    subset = df_monthly[(df_monthly['Jurisdiction of Occurrence'] == state) & (df_monthly['Week Ending Date'].month == month.month) & (df_monthly['Week Ending Date'].year == month.year)]
    filename = f"{state}_{month.strftime('%Y-%m')}.csv"
    subset.to_csv(filename, index=False)
```

```python
state_month_combinations
```

```python
state_month_combinations = df_monthly[['Jurisdiction of Occurrence', 'Week Ending Date']].drop_duplicates()

for state, month in state_month_combinations.itertuples(index=False):
    subset = df_monthly[(df_monthly['Jurisdiction of Occurrence'] == state) & (df_monthly['Week Ending Date'].dt.month == month.month) & (df_monthly['Week Ending Date'].dt.year == month.year)]
    filename = f"./static/month/{state}_{month.strftime('%Y-%m')}.csv"
    subset.to_csv(filename, index=False)
```
